```python
[output forloop]
[<output> <forloop> <if cond>]
[<if out> <if> <else> <else out> <for loop>]
```

```python
l=[90, 65, 80, 50]
#='A', 'B', 'A', 'C'
# >75 A
# >65 B
# else C
out = ['A' if i>75 else 'B' if i>64 else 'C' for i in l]
out

out = ['A' if i>=75 \
       else 'B' if i>=65 \
       else 'C' \
       for i in l]
out
```
