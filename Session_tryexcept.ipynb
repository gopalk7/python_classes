```python
try:
    marks=eval(input('Enter your marks'))
    if marks>=75:
        print('A')
    elif marks>=60:
        print('B')
    elif marks>=45:
        print('C')
    elif marks>=35:
        print('D')
    else:
        print('fail')
except Exception as e:
    print(e)
    
```

```python
try:
    print('coffee is good')
except Exception as e:
    print('coffee is bad')
finally:
    print('Off the stove')
```
